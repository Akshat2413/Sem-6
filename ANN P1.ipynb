```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 linearly spaced values between -5 and 5
x = np.linspace(-5, 5, 100)

# Define activation functions and compute their values for input x
activation_functions = {
    # Sigmoid function: Maps input to a range between 0 and 1
    'Sigmoid': 1 / (1 + np.exp(-x)),  

    # Tanh function: Maps input to a range between -1 and 1
    'Tanh': np.tanh(x),  

    # ReLU function: Replaces negative values with 0 and keeps positive values unchanged
    'ReLU': np.maximum(0, x),  

    # Softmax function: Converts values into probabilities
    'Softmax': np.exp(x) / np.sum(np.exp(x))  
}

# Iterate to plot graph
for i, (name, y) in enumerate(activation_functions.items(), start=1):
    plt.figure(i)
    plt.plot(x, y, label=name)
    plt.xlabel('Input')
    plt.ylabel('Activation')
    plt.title(f'{name} Activation Function')
    plt.legend()
    plt.grid(True)

plt.show()
```
